## Dataset Overview

> The dataset contains variables such as duration of the trip, start station and end station, start time and end time timestamp, gender, birth years and user type. 

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

```

```python
# load in the dataset into a pandas dataframe
df_clean=pd.read_csv('Clean_data_bikes_share.csv')
```

> Note that the above cells have been set as "Skip"-type slides. That means

## Age Distribution of the Users

  #### Age is skewed with high number between 25 and 40 years. There are some outliers in this variable,120years.
  #### The highest frequency being between 30 and 40.

```python
def x_y_t(xlabel,ylabel,title):
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
color=sb.color_palette()[0]
#visualize the Age distribution
bins=np.arange(15,df_clean['Age'].max()+5,5)
plt.hist(data=df_clean,x='Age',color=color,bins=bins)
x_y_t('Age','Frequency','Age Distribution')
plt.show()
```

## Gender Distribution

#### Most members in the dataset are Male, with a huge difference from the female

```python
#create a pie chart to better illustrate the difference in memmber gender as relative frequency
sorted_counts=df_clean['member_gender'].value_counts()
my_labels=['Male',"Female",'Other']
plt.pie(sorted_counts,labels=my_labels,startangle=90,counterclock=False)
plt.axis('square')
plt.title('Member Gender')
plt.show()
```

## Bike Sharing

#### There are few people sharing their bike trips

```python
#create a pie chart for the bike sharing response
sorted_counts=df_clean['bike_share_for_all_trip'].value_counts()
my_labels=df_clean['bike_share_for_all_trip'].unique()
plt.pie(sorted_counts,labels=my_labels,startangle=90,counterclock=False)
plt.axis('square')

print(df_clean['bike_share_for_all_trip'].value_counts())
plt.title("Bike sharing for Trips")
plt.show()
```

### Days Distribution
#### Thursday is the day of the week with the highest number of start time and end time

```python
order=['Monday','Tuesday',"Wednesday",'Thursday','Friday','Sartuday','Sunday']
sb.countplot(data=df_clean,x='StartDay_of_week',color=color,order=order)
plt.xticks(rotation=30)
x_y_t('Day of the Week','count','Start Days of the Week')
plt.show()
```
